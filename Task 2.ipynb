# Data Cleaning

```python
df.isnull().sum() # to check missing values
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.drop('Cabin',axis=1,inplace=True)
```

# Univariate Analysis

```python
#Numerical Variables
plt.figure(figsize=(12,10))
for i,col in enumerate(['Age','SibSp','Parch','Fare']):
    plt.subplot(2,2,i+1)
    sns.histplot(df[col],bins=20,kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
#categorical variables
plt.figure(figsize=(12,6))
for i,col in enumerate(['Pclass','Sex','Embarked']):
    plt.subplot(1,3,i+1)
    sns.countplot(x=col,data=df)
    plt.title(f'Count of Passengers by {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
# target variable
plt.figure(figsize=(8,5))
sns.countplot(x='Survived',data=df)
plt.title("Count of Passengers by Survival Status")
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
#Bivariate Analysis
sns.scatterplot(x='Age',y='Fare',data=df,hue="Pclass")
plt.title("Scatterplot of  Age vs Fare , colored by Passenger Class ")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title='Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Fare',data=df,hue='Survived')
plt.title("scatterplot of Age vs Fare, Sized by Survival Status")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title='Survived',loc='upper right')
plt.show()
```

```python
#Scatter plot of age vs Fare ,Faceted by Embarked Port
sns.relplot(x='Age',y='Fare',data=df,col='Embarked',hue='Pclass',kind='scatter')
plt.show()
```

```python
#barplot
fig,axes=plt.subplots(2,2,figsize=(15,12))
sns.countplot(x='Pclass',hue='Survived',data=df,ax=axes[0,0])
axes[0,0].set_title("Survived Vs Pclass")
sns.countplot(x='Sex',hue="Survived",data=df,ax=axes[0,1])
axes[0,1].set_title("Survived Vs Sex")
sns.countplot(x="Embarked",hue="Survived",data=df,ax=axes[1,0])
axes[1,0].set_title("Survived Vs Embarked ")
sns.countplot(x="SibSp",hue="Survived",data=df,ax=axes[1,1])
axes[1,1].set_title("Survived Vs SibSp")
plt.tight_layout()
plt.show()
```

```python
#Boxplots
fig,axes=plt.subplots(2,2,figsize=(15,12))
sns.boxplot(x='Survived',y='Age',data=df,ax=axes[0,0])
axes[0,0].set_title("Age Vs Survived")
sns.boxplot(x='Survived',y='Fare',data=df,ax=axes[0,1])
axes[0,1].set_title("Fare Vs Survived")
sns.boxplot(x='Pclass',y='Age',data=df,ax=axes[1,0])
axes[1,0].set_title("Age Vs Pclass")
sns.boxplot(x='Pclass',y='Fare',data=df,ax=axes[1,1])
axes[1,1].set_title("Fare Vs Pclass")
plt.tight_layout()
plt.show()
```

```python
# multivariate Analysis
sns.pairplot(df,vars=['Age',"Fare","Pclass","SibSp","Parch"],hue='Survived',diag_kind='kde')
plt.suptitle("Pair Plot of Titanic Dataset",y=1.02)
plt.show()
```

```python
num_col=['Age',"Fare","Pclass","SibSp","Parch","Survived"]
correlation_matrix=df[num_col].corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation matrix of Titanic Dataset")
plt.show()
```
